```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
n=48
d=4
# Lattice -> p = 0
REG = nx.watts_strogatz_graph(n, d, 0, seed=None)
# Visualization
nx.draw_circular(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n=48
M = 150
RGER = nx.gnm_random_graph(n, M)
# Visualization of RGER
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n=48
R=0.35

RGG = nx.random_geometric_graph(n, R, dim=2) 
pos = nx.get_node_attributes(RGG, 'pos')
nx.draw_networkx_nodes(RGG, pos, node_size=82.5, node_color="r")
nx.draw_networkx_edges(RGG, pos, width=0.06, edge_color="b")
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n=48
d=4
SF = nx.barabasi_albert_graph(n, d)
nx.draw_circular(SF)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n=48
d=4
p=0.25
SW = nx.watts_strogatz_graph(n, d, p)
nx.draw_circular(SW)
```

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

#### <font color='blue'>B2.1</font>

```python
def compute_path_lengths(G):
    return nx.all_pairs_shortest_path_length(G)

REG_path_lengths = list(compute_path_lengths(REG))
RGER_path_lengths = list(compute_path_lengths(RGER))
RGG_path_lengths = list(compute_path_lengths(RGG))
SF_path_lengths = list(compute_path_lengths(SF))
SW_path_lengths = list(compute_path_lengths(SW))

def getLengths(G_path_lengths):
    lengths = []
    for item in G_path_lengths:
        item_dict = item[1]
        for value in item_dict.values():
            lengths.append(value)
            
    return lengths

REG_lengths = getLengths(REG_path_lengths)
RGER_lengths = getLengths(RGER_path_lengths)
RGG_lengths = getLengths(RGG_path_lengths)
SF_lengths = getLengths(SF_path_lengths)
SW_lengths = getLengths(SW_path_lengths)

print("Mean path length REG: " + str(np.mean(REG_lengths)))
print("Variance of path length REG: " + str(np.var(REG_lengths)))
print()

print("Mean path length RGER: " + str(np.mean(RGER_lengths)))
print("Variance of path length RGER: " + str(np.var(RGER_lengths)))
print()

print("Mean path length RGG: " + str(np.mean(RGG_lengths)))
print("Variance of path length RGG: " + str(np.var(RGG_lengths)))
print()

print("Mean path length SF: " + str(np.mean(SF_lengths)))
print("Variance of path length SF: " + str(np.var(SF_lengths)))
print()

print("Mean path length SW: " + str(np.mean(SW_lengths)))
print("Variance of path length SW: " + str(np.var(SW_lengths)))
```

#### <font color='magenta'>Σχολιασμός</font>

Αρχικά παρατηρούμέ ότι ο γράφος **REG** έχει το μεγαλύτερο μέσο μήκος μονοπατιού. Αυτό το περιμέναμε, καθώς κάθε κόμβος είναι συνδεδεμένος με d κόμβους. Επίσης το variance του γράφου είναι αυξημένο, καθώς οι γείτονες κόμβοι είναι συνδεδεμένοι με μήκος μονοπατιού ίσο με 1, το οποίο αυξάνεται προοδευτικά μέχρι το $ n \over 4 $. Τα αποτελέσματα για τον **RGER** γράφο είναι επίσης τα αναμενόμενα. Το μικρό average path είναι αποτέλεσμα της τυχαιότητας της δημιουργίας των ακμών που ακολουθεί ομοιόμορφη κατανομή. Ο γράφος **RGG** έχει μικρό average path σε σχέση με αυτό που αναμέναμε, εξ'αιτίας των cliques που δημιουργούνται, καθώς κάθε μέλος μίας κλίκας απέχει αρκετά από ένα μέλος μίας άλλης κλίκας.  Στους **SW** γράφους, το αποτέλεσμα του average path είναι μεγάλο σε σχέση με αυτό που περιμέναμε. Ωστόσο αυτό μπορεί να δικαιολογηθεί εδώ, εξ'αιτίας της μικρής πιθανότητας rewiring που ορίζουμε, κάτι το οποίο οδηγεί σε μικρότερα shortcuts. Η απόκλιση του **SF** από το RGER όσον αφορά το average path είναι μεγαλύτερη.

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
    diameter = nx.diameter(G)
    result = 0
    len_result = 0
    final_result = []
    i = 0
    for node in sorted(G.nodes()):
        start_end_nodes = sorted([(node, k) for k,v in nx.shortest_path_length(G, node).items()
                                            if v == diameter])
        if len(start_end_nodes) >len_result:
            len_result = len(start_end_nodes)
            result = node
    for node in sorted(G.nodes()):
        start_end_nodes = sorted([(node, k) for k,v in nx.shortest_path_length(G, node).items()
                                            if v == diameter])
        if len(start_end_nodes) == len_result:
            final_result.append( node)
            i+=1
    return [final_result,len_result]
    
            
```

```python
x = find_nodes(REG)
print("REG: The nodes",x[0],"are connected with",x[1], "nodes with length of min path equal to the diameter.")
x = find_nodes(RGER)
print("RGER: The node",x[0],"is connected with",x[1], "nodes with length of min path equal to the diameter.")
x = find_nodes(RGG)
print("RGG: The node",x[0],"is connected with",x[1], "nodes with length of min path equal to the diameter.")
x = find_nodes(SF)
print("SF: The nodes",x[0],"are connected with",x[1], "nodes with length of min path equal to the diameter.")
x = find_nodes(SW)
print("SW: The node",x[0],"is connected with",x[1], "nodes with length of min path equal to the diameter.")

# να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων 
# με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ΄ίσο με τη διάμετρο. 
```

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G):
    lamda = nx.adjacency_spectrum(G).max()
    alpha = 1/lamda - 0.01
    return nx.katz_centrality(G,alpha)
```

```python
REG_katz = find_katz(REG)
RGER_katz = find_katz(RGER)
RGG_katz = find_katz(RGG)
SF_katz = find_katz(SF)
SW_katz = find_katz(SW)

Y = np.array(range(n))/float(n)

plt.figure(figsize=(7,7))
plt.title("Cumulative Katz centralities distribution")
plt.xlabel("Katz centrality ")
plt.ylabel("cdf(x)")
plt.plot(np.sort(np.real(np.array(list(REG_katz.values())))), Y ,color="red", label="REG")
plt.plot(np.sort(np.real(np.array(list(RGER_katz.values())))), Y ,color="blue", label="RGER")
plt.plot(np.sort(np.real(np.array(list(RGG_katz.values())))), Y ,color = "yellow", label="RGG")
plt.plot(np.sort(np.real(np.array(list(SF_katz.values())))), Y , color="green", label="SF")
plt.plot(np.sort(np.real(np.array(list(SW_katz.values())))), Y , color="black", label="SW")
plt.legend(loc="best")
```

```python
# REG connectivity

y = np.zeros(10)

for n in [100,200]:
  for d in range (2,11,2):
    foer i in range (100):
      REG_= nx.watts_strogatz_graph(n, d, 0)
      if nx.is_connected(REG_):
        if n==100: 
          y[d//2 -1] = y[d//2 -1] + 1
        else:
          y[d//2 +4] = y[d//2 +4] + 1

plt.plot(list(range(2,12,2)),y[0:5]/100, color='m', label="100 nodes")
plt.plot(list(range(2,12,2)),y[5:10]/100, color='b', label="200 nodes")
plt.title("Connectivity REG")
plt.ylabel("Connectivity percentage" )
plt.xlabel("Degree")
plt.legend()
plt.show()
```

Το αποτέλεσμα του κανονικού γράφου είναι το αναμενόμενο. Δηλαδή, παραμένει πάντα συνεκτικός, καθώς από τον ορισμό του κάθε κόμβος συνδέεται με τους d γείτονες του.

```python
# RGER connectivity
y = np.zeros(16)

for n in [100,200]:
  for M in range (100,801,100):
    for i in range (0,100):
        RGER_= nx.gnm_random_graph(n, M) 
        if nx.is_connected(RGER_):
          if n==100: 
            y[M//100 -1] = y[M//100 -1] + 1
          else:
            y[M//100 +7] = y[M//100 +7] + 1

#plot example:
plt.plot(list(range(100,900,100)),y[0:8]/100, color='m', label="100 nodes")
plt.plot(list(range(100,900,100)),y[8:16]/100, color='b', label="200 nodes")
plt.title("Connectivity RG Erdos-Renyi")
plt.ylabel("Connectivity percentage")
plt.xlabel("M")
plt.legend()
plt.show()
```

Στον RGER γράφο για μικρές τιμές του M η συνεκτικότητα είναι πολύ μικρή,γιατί για $Μ \leq n$, είναι αδύνατο ο γράφος να είναι συνδεδεμένος, αφού οι ακμές είναι λιγότερες από τις κορυφές. Γενικά, όσο αυξάνεται το M, τόσο αυξάνεται και η συνεκτικότητα, αλλά με διαφορετικό ρυθμό για τους 100 και 200 κόμβους. Παρατηρούμε επίσης, ότι στην περίπτωση με τους 200 κόμβους η συνεκτικότητα δεν φτάνει ποτέ στο 100%. 

```python
# RGG connectivity
y = np.zeros(18)
for n in [100,200]:
  for p in range (1,10, 1):
    for i in range (0,100):
      RG_gil_= nx.gnp_random_graph(n,p/100)
      if nx.is_connected(RG_gil_):
        if n==100:
          y[p -1] = y[p -1] + 1
        else:
          y[p +8] = y[p +8] + 1

#plot example:
plt.plot(np.arange(0.1, 1.0, 0.1),y[0:9]/100, color='m' , label="100 nodes")
plt.plot(np.arange(0.1, 1.0, 0.1),y[9:18]/100, color='b' , label="200 nodes")
plt.title("Connectivity RG Gilbert")
plt.ylabel("Connectivity percentage")
plt.xlabel("p")
plt.legend()
plt.show()
```

Παρατηρούμε ότι ο γράφος RG Gilbert, είναι πάντα συνεκτικός για κάθε p και ανεξάρτητα από τον αριθμό των κόμβων.

```python
# RGG connectivity
y = np.zeros(20)
L = 1000

for n in [100,200]:
  for R in range (25,251, 25):
    for i in range(0,100):
      RGG_temp= nx.random_geometric_graph(n, R/1000, 2)
      if nx.is_connected(RGG_temp):
        if n==100:
          y[R//25 -1] = y[R//25 -1] + 1
        else:
          y[R//25 +9] = y[R//25 +9] + 1

#plot example:
plt.plot(list(range(25,251, 25)),y[0:10]/100, color='b', label="100 nodes")
plt.plot(list(range(25,251, 25)),y[10:20]/100, color='r', label="200 nodes")
plt.title("Connectivity percentage for RGG")
plt.ylabel("Connectivity percentage")
plt.xlabel("R")
plt.legend()
plt.show()
```

Στην περίπτωση του RGG, παρατηρούμε ότι υπάρχει threshold για κάποια τιμή της ακτίνας R, η οποία είναι διαφορετική για διαφορετικό αριθμό κόμβων.

```python
# SF connectivity
y = np.zeros(10)

for n in [100,200]:
  for d in range (2,11, 2):
    for i in range (0,100):
      SF_ = nx.barabasi_albert_graph(n,d)
      if nx.is_connected(SF_):
        if n==100:
          y[d//2 - 1 ] = y[d//2 -1] + 1
        else:
          y[d//2 +4 ] = y[d//2 +4] + 1

#plot example:
plt.plot(list(range(2,11, 2)),y[0:5]/100, color='b', label="100 nodes")
plt.plot(list(range(2,11, 2)),y[5:10]/100, color='r', label="200 nodes")
plt.title("Connectivity percentage for SF")
plt.ylabel("Connectivity percentage")
plt.xlabel("d")
plt.legend(loc="best")
plt.show()
```

Παρατηρούμε ότι ανεξαρτήτως παραμέτρων, ο SF παραμένει πάντα συνεκτικός. Αυτό οφείλται στον τρόπο δημιουργίας των SF γράφων. Πολλοί κόμβοι είναι συνδεδεμένοι με τους δημοφιλείς κόμβους, κάτι που οδηγεί σε συνεκτικότητα.

```python
# SW connectivity
y = np.zeros((10,7))

for n in [100,200]:
  for d in range (2,11, 2):
    for g in range (1,8):
      for i in range (0,100):
        SW_ = nx.watts_strogatz_graph(n,d,g/10)
        if nx.is_connected(SW_):
          if n==100:
            y[d//2 -1 , g-1] = y[d//2 -1, g-1] + 1
          else:
            y[d//2 +4 , g-1] = y[d//2 +4, g-1] + 1

#plot example:
for g in range (0,7):
  plt.figure()
  plt.plot(list(range(2,11, 2)),y[:5, g]/100, color='b', label="100 nodes")
  plt.plot(list(range(2,11, 2)),y[5:, g]/100, color='r', label="200 nodes")
  title = "Connectivity SW (p=" + str((g+1)/10) + ")"
  plt.title(title)
  plt.ylabel("Connectivity percentage")
  plt.xlabel("d")
  plt.legend()
  plt.show()
```

Παρατηρούμε ότι υπάρχει threshold για τον SW γράφο. Για κάθε p, όταν για το d ισχύει ότι $ d>4 $ ο γράφος είναι συνεχώς συνεκτικός. Ωστόσο, καθώς ο SW προκύπτει από rewiring ακμών στον Regular Graph περιμέναμε ότι θα ήταν συνεκτικός και για $ d<4 $.

### E) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

##### Μεταβάλλοντας την παράμετρο $g_p$ από $0$ έως $1$ με βήμα $0.1$της συνάρτησης που κατασκευάζει τοπολογίες SW μεγέθους $n=30x$ ( αντικαταστήστε το x με το τελευταίο ψηφίο του ΑΜ) με βαθμό αρχικού πλέγματος $d=4$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
random.seed(seedno)
n=308
d=4

def clustering_coefficient(G):
    return list(nx.clustering(G).values())

def compute_path_lengths(G):
    length = list(nx.all_pairs_shortest_path_length(G))
    dictlist = list()
    for item in length:
        temp = list(item[1].values())
        dictlist.append(temp)
    return dictlist

for gp in np.arange(0, 1.1, 0.1): 
    SW = nx.watts_strogatz_graph(n, d, gp)
    SW_path_lengths = list(compute_path_lengths(SW))
    SW_clustering_coeff = clustering_coefficient(SW)

    print("SW Gp = %f" %gp)
    print("Mean path length SW: %f" %np.mean(SW_path_lengths))
    print("Mean clustering coefficient SW: %f" %np.mean(SW_clustering_coeff))
    print('----------------------------------------------------------')
```

#### <font color='magenta'>Σχολιασμός</font>

Παρατηρούμε ότι για $g_p$ = 0 ο γράφος τα ίδια χαρακτηριστικά με τον REG γράφο, γιατί πρακτικά είναι ένας REG γράφος. Όταν $g_p$ = 1 έχουμε πρακτικά ένα τυχαίο γράφο, καθώς έχει σχετικά μικρό  Clustering Coefficient, αλλα και σχετικά μικρό μήκος μονοπατιού. Όσον αφορά τις ενδιάμεσες τιμές παρατηρούμε πτώση στο μέσο μήκος μονοπατιού και clustering coefficient, επειδή όσο το $g_p$ αυξάνεται, τόσο περισσότερες από τις ακμές που συνέδεαν κοντινούς κόμβους συνδέονται πιο μακρυά.
